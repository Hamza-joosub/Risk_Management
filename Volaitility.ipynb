```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm, trange
```

```python
def find_volatility(Ticker, Start, End):
    Ticker = Ticker
    start = Start
    end = End
    time_period = 30
    rawData = yf.download(tickers=Ticker,start=start, end =end)
    quotes = pd.DataFrame(rawData["Close"])
    ema = pd.DataFrame(quotes).ewm(alpha = 0.25,min_periods=time_period).mean()
    ema["quote"] = quotes["Close"]
    ema["Diffirence"] = abs(ema["Close"] - quotes["Close"])
    ema["Close"].plot()
    rawData["Close"].plot()
    number = ema["Diffirence"].mean()
    plt.title(f"{Ticker}: {number}")
    return ema["Diffirence"].mean()
find_volatility(Ticker = "WRK", Start = "2021-01-01", End = "2023-12-19")
```
