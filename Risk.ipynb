```python
import yfinance as yf
from pandas import MultiIndex, Int16Dtype
from datetime import datetime, timedelta
import datetime
from datetime import *
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
date_end = "2023-10-06"
date_start = "2018-01-31"
interval = "1mo"
StockList = ["AVGO", "AMZN", "ULTA"]
```

```python
tickerSymbol1 = 'AVGO'
data1 = yf.Ticker(tickerSymbol1)
raw1 = data1.history(interval= interval,start = date_start, end = date_end)
stock1 = raw1["Close"]
print(stock1.shape)
```

```python
tickerSymbol2 = 'AMZN'
data2 = yf.Ticker(tickerSymbol2)
raw2 = data2.history(interval=interval,start = date_start, end = date_end)
stock2 = raw2["Close"]
print(stock2.shape)
stock2
```

```python
tickerSymbol3 = 'ULTA'
data3 = yf.Ticker(tickerSymbol3)
raw3 = data3.history(interval=interval,start = date_start, end = date_end)
stock3 = raw3["Close"]
print(stock3.shape)
```

```python
df = pd.DataFrame()
df["AVGO"] = stock1
df["AMZN"] = stock2
df["ULTA"] = stock3
matrix = df.corr()
```

```python
sns.heatmap(matrix,annot = True,cmap= "RdYlGn")
```

```python
corr = np.corrcoef((stock1, stock2, stock3))
print(corr)
plt.figure(figsize=(13, 8))
sns.heatmap(corr, annot = True,cmap= "RdYlGn")
plt.figure()
```
